# 분류 데이터 시각화

```python
# 필요한 라이브러리 가져오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

```python
# 데이터 불러오기
iris = datasets.load_iris()
iris
```

```python
# 데이터에 대한 정보 출력
print(iris.DESCR)
print(iris.keys())
print(iris.data)
print(iris.target)
print(iris.feature_names)
```

```python
# 데이터셋 생성
df = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length','petal_width'])
df['species'] = iris.target
df.head()
```

```python
# 범주를 문자형으로 변환
df['species'].replace(0,'setosa', inplace=True)
df['species'].replace(1,'versicolor', inplace=True)
df['species'].replace(2,'virginica', inplace=True)
df.head()
```

```python
# 데이터 요약
print('데이터 사이즈:', df.shape)
print('상위 5개 데이터 출력:\n', df.head(20))
print('통계요약:\n', df.describe())
tab = df.groupby('species').size()
pct = (tab / tab.sum())*100
tab = pd.concat([tab, pct], axis=1)
tab.columns = ['freq', 'percentage']
print('빈도표:\n', tab)
```

# 단변량 분석

```python
# 붓꽃 종류 시각화 (matplotlib)
plt.bar(tab.index, tab.freq)
plt.show()
```

```python
# 붓꽃 종류 시각화 (seaborn)
sns.countplot(df['species'])
plt.show()
```

```python
# 속성에 따른 박스플랏 (matplotlib)
df.boxplot()
plt.show()
```

```python
# 속성에 따른 박스플랏 (seaborn)
fig, axes = plt.subplots(ncols=4)
for i in range(4):
    sns.boxplot(data=df, y=df.columns[i], ax=axes[i])
plt.tight_layout()
plt.show()
```

```python
# 속성에 따른 히스토그램
df.hist()
plt.show()
```

# 다변량 분석

```python
# 산점도
pd.plotting.scatter_matrix(df)
plt.show()
```

```python
# 꽃종류에 따른 산점도
sns.pairplot(df, hue='species')
plt.show()
```

```python
# 속성과 꽃종류에 따른 박스플랏
fig, axes = plt.subplots(2,2)
sns.boxplot(x='species', y='sepal_length', data=df, ax=axes[0,0])
sns.boxplot(x='species', y='sepal_width', data=df, ax=axes[0,1])
sns.boxplot(x='species', y='petal_length', data=df, ax=axes[1,0])
sns.boxplot(x='species', y='petal_width', data=df, ax=axes[1,1])
plt.tight_layout()
plt.show()
```

```python
# 속성과 꽃종류에 따른 밀도그래프
fig, axes = plt.subplots(2,2)
sns.kdeplot(x='sepal_length', hue='species', data=df, ax=axes[0,0], legend=None)
sns.kdeplot(x='sepal_width', hue='species', data=df, ax=axes[0,1], legend=None)
sns.kdeplot(x='petal_length', hue='species', data=df, ax=axes[1,0], legend=None)
sns.kdeplot(x='petal_width', hue='species', data=df, ax=axes[1,1])
plt.tight_layout()
plt.show()
```
